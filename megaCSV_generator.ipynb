```python
df = pd.DataFrame();
for i in range(0,981,20):
    path = './data/f{}t{}/cars.txt'.format(i,i+19);
    #print(path)
    df1 = pd.read_table(path, sep='\t', header=None);
    df = pd.concat([df, df1], axis=0);
df = df.reset_index(drop=True);
df.columns = ['Year_Make_Model','Price','Mileage','Body','Color','Engine'];

print("-"*10 + " Raw Data " + "-"*10)
print(df.shape)

print("-"*10 + " Drop $Price == N/A " + "-"*10)
df = df.loc[df.Price.str.startswith('$')]
print(df.shape)

print("-"*10 + " Drop $Mileage == N/A " + "-"*10)
df = df.loc[df.Mileage!='N/A miles']
print(df.shape)


df.head(10)

print("-"*50)

# Column Year_Make_Model splits into three columns: Year, Make, Model
df['Year'] = \
list(map(lambda str4: str4[0],\
         list( map( lambda strLong: strLong.split(sep=' '), df.Year_Make_Model ) )));
df['Year'] = pd.to_numeric(df.Year);
df['Make'] = \
list(map(lambda str4: str4[1],\
         list( map( lambda strLong: strLong.split(sep=' '), df.Year_Make_Model ) )));
df['Model'] = \
list(map(lambda str4: str4[2],\
         list( map( lambda strLong: strLong.split(sep=' '), df.Year_Make_Model ) )));

print("-"*10 + " Drop $Make == Land " + "-"*10)
print('# Land Rover Range Rover:')
print('Not worth the efforts to go extra hours to fix this!')
# Land Rover Range Rover ## not worth the efforts to go extra miles to fix this
df = df.loc[df.Make!='Land']
print(df.shape)

df.head()
```

```python
# Price to numerics
df.Price = list(map(lambda str0: str0[1:].replace(',',''), df.Price));
df.Price = pd.to_numeric(df.Price);

# Mileage to numerics
df.Mileage = list(map(lambda str0: str0.replace(' miles', '').replace(',',''), df.Mileage));
df.Mileage = pd.to_numeric(df.Mileage);

df = df[['Year','Make','Model','Mileage','Price','Body','Color','Engine']];
df.head()
```
